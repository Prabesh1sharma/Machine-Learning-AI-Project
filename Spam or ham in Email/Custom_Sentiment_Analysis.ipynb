```python
import pandas as pd
df=messages = pd.read_csv('SMSSpamCollection.txt', sep='\t',
                           names=["label", "message"])
df.head()
```

```python
df.shape
```

```python
X=list(df['message'])
```

```python
y=list(df['label'])
```

```python
y
```

```python
y=list(pd.get_dummies(y,drop_first=True)['spam'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
%pip install transformers
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
y_train
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))
```

```python
from transformers import TFDistilBertForSequenceClassification, TFTrainer, TFTrainingArguments
import tensorflow as tf
from transformers import Trainer

# Set up the distributed training strategy
# strategy = tf.distribute.MirroredStrategy()

# Define training arguments
training_args = TFTrainingArguments(
    output_dir="./results",          # output directory
    num_train_epochs=3,              # number of training epochs
    per_device_train_batch_size=4,  # batch size for training
    per_device_eval_batch_size=8,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir="./logs",            # directory for storing logs
    logging_steps=10
)

# # Instantiate the model outside the strategy scope
# with strategy.scope():
#     model = TFDistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

#     # Manually build the model
#     dummy_input = tf.constant([[0]*128])  # Adjust input size if necessary
#     model(dummy_input)

#     trainer = TFTrainer(
#         model=model,                         # the instantiated 🤗 Transformers model to be trained
#         args=training_args,                  # training arguments, defined above
#         train_dataset=train_dataset,         # training dataset
#         eval_dataset=test_dataset            # evaluation dataset
#     )

# # Start training
# trainer.train()
model = TFDistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased")

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer,
    
)

trainer.train()
```

```python
trainer.evaluate(test_dataset)
```
